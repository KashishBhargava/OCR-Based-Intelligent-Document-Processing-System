```python
# Image path
IMAGE_PATH = "patta_sample.png"

# Load image
img = cv2.imread(IMAGE_PATH)

if img is None:
    raise FileNotFoundError("Image not found. Check the file path.")

# Convert image to grayscale
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# OCR: Extract text from image
text = pytesseract.image_to_string(gray, lang="eng")

print("========== RAW OCR TEXT ==========")
print(text)


# ---------------- ENTITY EXTRACTION ---------------- #

# Name
name = re.search(
    r"(Mane|Name).*?(claimant|clainant).*?:\s*(.*)",
    text,
    re.IGNORECASE
)

# Spouse
spouse = re.search(
    r"(spouse).*?:\s*(.*)",
    text,
    re.IGNORECASE
)

# Father / Mother
father = re.search(
    r"(father|mother).*?:\s*(.*)",
    text,
    re.IGNORECASE
)

# Village
village = re.search(
    r"(village).*?:\s*(.*)",
    text,
    re.IGNORECASE
)

# District
district = re.search(
    r"(district|pistrict).*?:\s*(.*)",
    text,
    re.IGNORECASE
)

# Self cultivation land
cultivation = re.search(
    r"self.?cultivation.*?:\s*(.*)",
    text,
    re.IGNORECASE
)


# ---------------- CLEANED OUTPUT ---------------- #

extracted_data = {
    "Name": name.group(3).strip() if name else None,
    "Spouse": spouse.group(2).strip() if spouse else None,
    "Father/Mother": father.group(2).strip() if father else None,
    "Village": village.group(2).strip() if village else None,
    "District": district.group(2).strip() if district else None,
    "Self Cultivation Land": cultivation.group(1).strip() if cultivation else None
}

print("\n========== EXTRACTED ENTITIES ==========")

for key, value in extracted_data.items():
    print(f"{key}: {value}")


# ---------------- STRUCTURED DATAFRAME ---------------- #

df = pd.DataFrame([extracted_data])

print("\n========== FINAL STRUCTURED TABLE ==========")
print(df)
```
